```python
import soundfile as sf
from scipy import signal
import librosa

def butter_highpass(fs, cutoff, order):
        nyq = 0.5 * fs
        normal_cutoff = cutoff / nyq
        b, a = signal.butter(order, normal_cutoff, btype='high', analog=False)
        return b, a

b, a = butter_highpass(fs=16000, cutoff=30, order=5)

path = '/work3/dgro/VCTK-Corpus-0/wav48_silence_trimmed/p227/p227_335_mic2.flac'
x_fs = librosa.load(path, sr=16000),
x = x_fs[0][0]
#Remove drifting noise
y = signal.filtfilt(b, a, x)
# Add a little random noise for model roubstness
#wav = y * 0.96 + (prng.rand(y.shape[0])-0.5)*1e-06
```
